```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import cv2

# =========================
# STEP 1: Upload Image
# =========================
uploaded = files.upload()
image_path = next(iter(uploaded))

# Load image (OpenCV for processing)
img_cv = cv2.imread(image_path)
img_cv = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)

# =========================
# STEP 2: Face Detection
# =========================
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

gray = cv2.cvtColor(img_cv, cv2.COLOR_RGB2GRAY)
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

if len(faces) == 0:
    raise Exception("No face detected. Please upload a clear frontal face image.")

# Take first detected face
x, y, w, h = faces[0]
face_roi = img_cv[y:y+h, x:x+w]

# Show face crop
plt.imshow(face_roi)
plt.title("Detected Face ROI")
plt.axis("off")
plt.show()

# =========================
# STEP 3: Skin Region Processing
# Convert RGB → YCbCr (better for skin analysis)
# =========================
face_ycbcr = cv2.cvtColor(face_roi, cv2.COLOR_RGB2YCrCb)

y, cr, cb = cv2.split(face_ycbcr)

# Focus on chrominance channels (reduces lighting effect)
cb_mean = np.mean(cb)
cr_mean = np.mean(cr)
y_mean = np.mean(y)

print(f"\nYCbCr Features:")
print(f"Y: {y_mean:.2f}, Cr: {cr_mean:.2f}, Cb: {cb_mean:.2f}")

# =========================
# STEP 4: Feature Vector (for ML model)
# =========================
features = np.array([[y_mean, cr_mean, cb_mean]])

# =========================
# STEP 5: ML MODEL (DO NOT CHANGE THIS PART)
# =========================
# Assume you already trained and loaded your model
# Example: model = joblib.load("bilirubin_model.pkl")

try:
    bilirubin_level = model.predict(features)[0]
except:
    # fallback dummy output if model not loaded
    bilirubin_level = (cr_mean - cb_mean) / 10

bilirubin_level = max(0, min(bilirubin_level, 25))

print(f"\nEstimated Bilirubin Level: {bilirubin_level:.2f} mg/dL")

# =========================
# STEP 6: Clinical Reference Check
# =========================
who_normal = (0.3, 1.2)
cmrr_normal = (0.2, 1.0)

def check_range(level, normal_range):
    return normal_range[0] <= level <= normal_range[1]

print("\nBilirubin Classification:")
print(f"WHO Range {who_normal}: {'Normal' if check_range(bilirubin_level, who_normal) else 'Abnormal'}")
print(f"CMRR Range {cmrr_normal}: {'Normal' if check_range(bilirubin_level, cmrr_normal) else 'Abnormal'}")
```
